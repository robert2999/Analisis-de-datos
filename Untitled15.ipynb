```python
import pandas as pd
```

```python
data = pd.read_excel("datos.xlsx", skiprows=3)
data = data.rename(columns={"Unnamed: 2":"FECHA", "Unnamed: 3":"TIENDA", "Unnamed: 4":"MONEDA", "Unnamed: 5":"INGRESOS", "Unnamed: 6":"COSTO"})
data
```

```python
data = data.drop(data.index[0])
data
```

```python
total_dolares=data['INGRESOS'].sum()
total_dolares
```

```python
total_soles=data['COSTO'].sum()*3.75
total_soles
```

```python
import matplotlib.pyplot as plt
```

```python
data['MES'] = pd.to_datetime(data['FECHA']).dt.month
data
```

```python
monthly_cost = data.groupby('MES')['COSTO'].sum()

plt.plot(monthly_cost.index, monthly_cost.values)
plt.xlabel('MES')
plt.ylabel('Total COSTO')
plt.title('Total COSTO by Month')
plt.xticks(monthly_cost.index) # Ensure all months are shown on the x-axis
plt.show()
```

```python
tienda_1L_ingresos = data[data['TIENDA'] == 'Tienda 1L']['INGRESOS'].sum()
tienda_2P_ingresos = data[data['TIENDA'] == 'Tienda 2P']['INGRESOS'].sum()
tienda_1P_ingresos = data[data['TIENDA'] == 'Tienda 1P']['INGRESOS'].sum()

print(f"Total ingresos para Tienda 1L: {tienda_1L_ingresos} dolares")
print(f"Total ingresos para Tienda 2P: {tienda_2P_ingresos} dolares ")
print(f"Total ingresos para Tienda 1P: {tienda_1P_ingresos} dolares ")
```
