```python
plt.style.use('ggplot')
fig,ax=plt.subplots(figsize=(15,10))
ax.plot(data['Fecha'],data['Elon Mosk'],label='Elon Mosk',lw=2,color='red')
ax.plot(data['Fecha'],data['Bill Gate'],label='Bill Gate',lw=2,color='blue',linestyle='-.')
ax.legend(shadow=True,prop={'family':'Serif','size':14})
ax.set_ylabel("USD",fontname='Serif',fontsize=15)
ax.set_xlabel("Tiempo",fontname="Serif",fontsize=15)
ax.set_title("Elon Musk vs Bill Gate",fontname="Serif",fontsize=25)
plt.show()
```

```python
r=pd.read_csv('compra+curso.csv',sep=';')
r
```

```python
plt.style.use('ggplot')
fig,ax=plt.subplots(figsize=(13,10))
ax.plot(r['Compras_del_curso'],r['calificaciones'],label='calificaciones',marker='o',lw=0,color='red',markersize=10)
ax.set_ylabel("calificaciones",fontname='Serif',fontsize=15)
ax.set_xlabel("Compras del curso",fontname="Serif",fontsize=15)
ax.set_title("Compras del curso vs calificaciones",fontname="Serif",fontsize=25)
plt.show()
```
