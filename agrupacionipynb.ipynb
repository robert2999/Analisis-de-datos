```python
datos={'Nombre y apellido':['Esteban','Dennise','Alvaro','Valentina','Damian','Celia','Soledad','Gabriel','Lucia','Sofia','Dennise'],'Edad':[54,24,33,41,37,19,28,54,18,33,37],
       'Pais':['Mexico','España','Honduras','Colombia','España','Brasil','Estados Unidos','China','Argentina','Argentina','Argentina']
       ,'Automovil':['Si','No','No','Si','Si','No','No','No','Si','No','Si']}


dat=pd.DataFrame(datos)
dat
```

Vamos a filtrar del dataframe y que devuelva todos los valores que cumplan que sean del pais de españa , pero a la vaez que tengan automovil

```python
filtrar=dat[((dat.Pais=='España')|(dat.Pais=='Argentina'))&(dat.Automovil=='Si')]
filtrar
```

```python
filtrar=dat[((dat.Pais=='España')|(dat.Pais=='Argentina'))]
filtrar
```

```python
agrupamos=filtrar.groupby(['Pais','Automovil']).size()
agrupamos
```

Pais      Automovil
Argentina No           1   ← 1 persona en Argentina sin automóvil
          Si           2   ← 2 personas en Argentina con automóvil
España    No           1   ← 1 persona en España sin automóvil
          Si           1   ← 1 persona en España con automóvil


```python
unstack=agrupamos.unstack()
unstack
```
